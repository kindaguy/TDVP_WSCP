```python
import numpy as np
import scipy as sp
import math
from tedopa import _recurrence_coefficients as rc
import matplotlib.pyplot as plt
```

```python
# Parameters
ss = 1.
cutoff = 1.
coupling  = 1.
temp = 77.
```

```python
jj=  lambda x: coupling / np.power(np.sign(x), ss - 1) * np.power(x, ss) * np.exp(-1 * np.fabs(x) / cutoff)*0.5*(1.+ 1./np.tanh(0.5*x*1.4387773/temp))
```

```python
alphas,betas = rc.recurrenceCoefficients(700,-10,10,jj,1.)
```

```python
omegas = np.array(alphas)
coups = np.sqrt(betas)
ts = np.sqrt(np.array(betas[1::]))
c0 = np.sqrt(betas[0])
```

```python
plt.plot(omegas)
```

```python
plt.plot(ts)
plt.ylim((4.99,5.01))
plt.xlim((100,120))
```
